```c++
# Diagrama de Flujo - Explicación

Este es un **diagrama de flujo** con **condicionales anidados** que calcula y muestra límites de una matriz según si el orden es par o impar. A continuación te explico cada elemento:

## Elementos del Diagrama

| Símbolo | Color | Significado | Descripción |
|---------|-------|-------------|-------------|
| 🔵 **Óvalo** | Morado | Principal | Inicio del programa |
| 🟨 **Rectángulo doble** | Amarillo | Entero i, j, limite, n | Declaración de variables |
| 🟩 **Paralelogramo** | Verde | Salida "Diga el valor de n" | Mensaje al usuario |
| 🔷 **Paralelogramo** | Azul | Entrada n | Leer valor de n |
| 🟨 **Rectángulo** | Amarillo | i = 1 | Inicializar contador |
| 🟨 **Rectángulo** | Amarillo | limite = Int(n/2+0.5) | Calcular límite inicial (redondeo) |
| 🟧 **Rombo** | Naranja | i &lt;= limite | **Bucle principal** - condición |
| 🟨 **Rectángulo** | Amarillo | j = 1 | Inicializar contador j |
| 🟧 **Rombo** | Naranja | n%2 == 0 | **Condición clave:** ¿n es par? |
| 🟩 **Paralelogramo** | Verde | Salida "Procesar " & i | Mostrar iteración (rama par) |
| 🟨 **Rectángulo** | Amarillo | limite = n | Asignar límite = n |
| 🟩 **Paralelogramo** | Verde | Salida "Para 'i=1' Su limite calculado es: " & limite | Mostrar límite calculado |
| 🟨 **Rectángulo** | Amarillo | limite = 2*(limite-i+1)-1 | **Fórmula A:** Calcular nuevo límite (par) |
| 🟩 **Paralelogramo** | Verde | Salida "Para 'i=2' Su limite calculado es: " & limite | Mostrar segundo límite |
| 🟨 **Rectángulo** | Amarillo | limite = n-2 | **Fórmula B:** Ajustar límite (par) |
| 🟩 **Paralelogramo** | Verde | Salida "Para 'i=3' Su limite calculado es: " & limite | Mostrar tercer límite |
| 🟨 **Rectángulo** | Amarillo | limite = n-2 | **Fórmula C:** Ajustar límite (par) |
| 🟩 **Paralelogramo** | Verde | Salida "Para 'i=4' Su limite calculado es: " & limite | Mostrar cuarto límite |
| 🟩 **Paralelogramo** | Verde | Salida "Procesar " & i | Mostrar iteración (rama impar) |
| 🟨 **Rectángulo** | Amarillo | limite = n | Asignar límite = n |
| 🟩 **Paralelogramo** | Verde | Salida "Para 'i=1' Su limite calculado es: " & limite | Mostrar límite inicial (impar) |
| 🟨 **Rectángulo** | Amarillo | limite = 2*(limite-i+1) | **Fórmula D:** Calcular nuevo límite (impar) |
| 🟩 **Paralelogramo** | Verde | Salida "Para 'i=2' Su limite calculado es: " & limite | Mostrar segundo límite (impar) |
| 🟨 **Rectángulo** | Amarillo | limite = n | **Fórmula E:** Restablecer límite (impar) |
| 🟩 **Paralelogramo** | Verde | Salida "Para 'i=3' Su limite calculado es: " & limite | Mostrar tercer límite (impar) |
| 🟨 **Rectángulo** | Amarillo | limite = n | **Fórmula F:** Restablecer límite (impar) |
| 🟩 **Paralelogramo** | Verde | Salida "Para 'i=4' Su limite calculado es: " & limite | Mostrar cuarto límite (impar) |
| 🟨 **Rectángulo** | Amarillo | i = i+1 | Incrementar contador |
| 🔵 **Óvalo** | Morado | Fin | Fin del programa |

## Fórmulas Matemáticas Clave

| Fórmula | Condición | Descripción |
|---------|-----------|-------------|
| `limite = Int(n/2 + 0.5)` | Inicial | Redondeo de n/2 al entero más cercano |
| `limite = 2*(limite-i+1)-1` | n par | Fórmula de ajuste para matrices pares |
| `limite = n-2` | n par | Decremento fijo de 2 |
| `limite = 2*(limite-i+1)` | n impar | Fórmula de ajuste para matrices impares |
| `limite = n` | n impar | Restablecimiento al valor original |

## Flujo del Algoritmo (Paso a Paso)

| Paso | Acción | Descripción |
|------|--------|-------------|
| 1 | Inicio | Comienza el programa |
| 2 | Declarar variables | i, j, limite, n |
| 3 | Pedir valor de n | Solicitar entrada al usuario |
| 4 | Leer n | Input del usuario |
| 5 | i = 1 | Inicializar contador |
| 6 | Calcular limite inicial | `Int(n/2 + 0.5)` |
| 7 | ¿i &lt;= limite? | Verificar condición del bucle |
| 8 | j = 1 | Inicializar contador interno |
| 9 | ¿n % 2 == 0? | **Bifurcación:** ¿n es par o impar? |
| 10 | Rama PAR: Mostrar "Procesar i" | Indicar iteración actual |
| 11 | limite = n | Asignar n como límite |
| 12 | Mostrar límite para i=1 | Salida informativa |
| 13 | Calcular límite con fórmula A | `2*(limite-i+1)-1` |
| 14 | Mostrar límite para i=2 | Salida informativa |
| 15 | Calcular límite con fórmula B | `n-2` |
| 16 | Mostrar límite para i=3 | Salida informativa |
| 17 | Calcular límite con fórmula C | `n-2` |
| 18 | Mostrar límite para i=4 | Salida informativa |
| 19 | Rama IMPAR: Mostrar "Procesar i" | Indicar iteración actual |
| 20 | limite = n | Asignar n como límite |
| 21 | Mostrar límite para i=1 | Salida informativa |
| 22 | Calcular límite con fórmula D | `2*(limite-i+1)` |
| 23 | Mostrar límite para i=2 | Salida informativa |
| 24 | Calcular límite con fórmula E | `n` |
| 25 | Mostrar límite para i=3 | Salida informativa |
| 26 | Calcular límite con fórmula F | `n` |
| 27 | Mostrar límite para i=4 | Salida informativa |
| 28 | i = i + 1 | Incrementar contador principal |
| 29 | Repetir 7-28 | Volver a verificar bucle |
| 30 | Fin | Programa termina |

## Comparación: n Par vs n Impar

### Si n es PAR (ejemplo: n = 4)

| Iteración | Fórmula aplicada | Cálculo | Resultado |
|-----------|------------------|---------|-----------|
| i=1 | limite = n | 4 | 4 |
| i=2 | 2*(limite-i+1)-1 | 2*(4-1+1)-1 = 2*4-1 | 7 |
| i=3 | n-2 | 4-2 | 2 |
| i=4 | n-2 | 4-2 | 2 |

### Si n es IMPAR (ejemplo: n = 5)

| Iteración | Fórmula aplicada | Cálculo | Resultado |
|-----------|------------------|---------|-----------|
| i=1 | limite = n | 5 | 5 |
| i=2 | 2*(limite-i+1) | 2*(5-1+1) = 2*5 | 10 |
| i=3 | n | 5 | 5 |
| i=4 | n | 5 | 5 |
```
