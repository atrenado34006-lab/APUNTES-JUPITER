```c++
Diagrama de Flujo: Generador de Patrón Numérico por Ciclos Externos
Descripción General
Este algoritmo genera un patrón numérico basado en ciclos externos, calculando valores que dependen del número de capa actual. Utiliza una fórmula matemática que involucra valor absoluto para determinar cuántos números imprimir en cada ciclo.
🔧 Variables
Table
Variable	Tipo	Descripción
i	Entero	Contador de ciclos externos (capas)
j	Entero	Contador interno para imprimir valores
Orden	Entero	Valor base ingresado por el usuario
valor	Entero	Cantidad de números a imprimir en cada ciclo
n	Entero	Tamaño calculado de la matriz/espiral
🔄 Flujo del Algoritmo
Markdown
Copy
Code
Preview
1. *INICIO* (Principal)

2. *DECLARACIÓN*: Entero i, j, Orden, valor, n

3. *SALIDA*: "Diga el Orden:"

4. *ENTRADA*: Leer Orden

5. *ASIGNACIÓN*: n = Orden * 2 - 1
   → Convierte el orden en un tamaño impar (1→1, 2→3, 3→5, etc.)

6. *ASIGNACIÓN*: i = 1

7. *CONDICIÓN*: ¿i <= n?
   
   ├─ *FALSO* → Ir a FIN
   
   └─ *VERDADERO*:
      │
      ├── *SALIDA*: "Ciclo externo: " & i
      │
      ├── *ASIGNACIÓN*: valor = n/2 - Abs(Int(n/2) - i) + 1
      │   → Fórmula que calcula la "profundidad" o posición simétrica
      │
      ├── *ASIGNACIÓN*: j = 1
      │
      ├── *CONDICIÓN*: ¿j <= valor?
      │   
      │   ├─ *FALSO* → Continuar al paso 8
      │   │
      │   └─ *VERDADERO*:
      │       ├── *SALIDA*: j
      │       ├── *ASIGNACIÓN*: j = j + 1
      │       └── *RETORNO* a condición j <= valor
      │
      ├── *ASIGNACIÓN*: i = i + 1
      │
      └── *RETORNO* a condición i <= n (paso 7)

8. *FIN*
```
