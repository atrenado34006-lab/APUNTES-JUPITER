```c++
# Diagrama de Flujo - Explicación

Este es un **diagrama de flujo** que simula el comportamiento de matrices cuadradas de diferentes órdenes. A continuación te explico cada elemento:

## Elementos del Diagrama

| Símbolo | Color | Significado | Descripción |
|---------|-------|-------------|-------------|
| 🔵 **Óvalo** | Morado | Principal | Inicio del programa |
| 🟨 **Rectángulo doble** | Amarillo | Entero OrdenMatriz, OrdenMaxima, NumMaxCiclosExt | Declaración de variables |
| 🟩 **Paralelogramo** | Verde | Salida "Orden de matriz maxima a ser probada" | Mensaje al usuario |
| 🔷 **Paralelogramo** | Azul | Entrada OrdenMaxima | Leer el orden máximo de matriz |
| 🟨 **Rectángulo** | Amarillo | OrdenMatriz = 1 | Inicializar contador |
| 🟧 **Rombo** | Naranja | OrdenMatriz &lt;= OrdenMaxima | **Condición del bucle** |
| 🟨 **Rectángulo** | Amarillo | NumMaxCiclosExt = Int(OrdenMatriz/2+0.5) | Calcular número máximo de ciclos externos |
| 🟩 **Paralelogramo** | Verde | Salida "Para una matriz cuadrada de orden" & OrdenMatriz & " = " & NumMaxCiclosExt & " ciclos externos maximo" | Mostrar resultado |
| 🟨 **Rectángulo** | Amarillo | OrdenMatriz = OrdenMatriz+1 | Incrementar contador |
| 🔵 **Óvalo** | Morado | Fin | Fin del programa |
```

```c++

Esta fórmula calcula el **número máximo de ciclos externos** para una matriz cuadrada de orden N. Es equivalente a redondear `OrdenMatriz / 2` al entero más cercano.

## Flujo del Algoritmo (Paso a Paso)

| Paso | Acción | Descripción |
|------|--------|-------------|
| 1 | Inicio | Comienza el programa |
| 2 | Declarar variables | OrdenMatriz, OrdenMaxima, NumMaxCiclosExt |
| 3 | Pedir "Orden de matriz maxima" | Mensaje al usuario |
| 4 | Leer OrdenMaxima | Input del usuario |
| 5 | OrdenMatriz = 1 | Inicializar contador en 1 |
| 6 | ¿OrdenMatriz <= OrdenMaxima? | Verificar condición del bucle |
| 7 | Calcular NumMaxCiclosExt | `Int(OrdenMatriz/2 + 0.5)` |
| 8 | Mostrar resultado | Información de la matriz actual |
| 9 | OrdenMatriz = OrdenMatriz + 1 | Incrementar contador |
| 10 | Repetir 6-9 | Volver a verificar condición |
| 11 | Fin | Programa termina |

## Ejemplo de Ejecución

Si el usuario ingresa `OrdenMaxima = 5`:

| Iteración | OrdenMatriz | Cálculo NumMaxCiclosExt | Resultado Mostrado |
|-----------|-------------|-------------------------|-------------------|
| 1 | 1 | Int(1/2 + 0.5) = Int(1) = **1** | "Para una matriz cuadrada de orden 1 = 1 ciclos externos maximo" |
| 2 | 2 | Int(2/2 + 0.5) = Int(1.5) = **1** | "Para una matriz cuadrada de orden 2 = 1 ciclos externos maximo" |
| 3 | 3 | Int(3/2 + 0.5) = Int(2) = **2** | "Para una matriz cuadrada de orden 3 = 2 ciclos externos maximo" |
| 4 | 4 | Int(4/2 + 0.5) = Int(2.5) = **2** | "Para una matriz cuadrada de orden 4 = 2 ciclos externos maximo" |
| 5 | 5 | Int(5/2 + 0.5) = Int(3) = **3** | "Para una matriz cuadrada de orden 5 = 3 ciclos externos maximo" |
```
