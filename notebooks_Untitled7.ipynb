```c++
# Diagrama de Flujo - Explicación

Este es un **diagrama de flujo complejo** con **bucles anidados y condicionales múltiples** que genera un patrón de matriz con coordenadas. A continuación te explico cada elemento:

## Elementos del Diagrama

| Símbolo | Color | Significado | Descripción |
|---------|-------|-------------|-------------|
| 🔵 **Óvalo** | Morado | Principal | Inicio del programa |
| 🟨 **Rectángulo doble** | Amarillo | Entero i, j, k, Orden, valor, n | Declaración de variables |
| 🟩 **Paralelogramo** | Verde | Salida "Diga el Orden:" | Mensaje al usuario |
| 🔷 **Paralelogramo** | Azul | Entrada Orden | Leer valor ingresado |
| 🟨 **Rectángulo** | Amarillo | n = Orden * 2 - 1 | Calcular dimensión de la matriz |
| 🟨 **Rectángulo** | Amarillo | k = 1 | Inicializar contador k (ciclos externos) |
| 🟨 **Rectángulo** | Amarillo | i = 1 | Inicializar contador i (filas) |
| 🟧 **Rombo** | Naranja | i &lt;= n | **Bucle externo** - condición de filas |
| 🟨 **Rectángulo** | Amarillo | valor = n/2-Abs(Int(n/2+0.5)-i)+1 | Calcular valor límite para columnas |
| 🟨 **Rectángulo** | Amarillo | j = 1 | Inicializar contador j (columnas) |
| 🟩 **Paralelogramo** | Verde | Salida "Ciclo externo:" & k | Mostrar número de ciclo externo |
| 🟧 **Rombo** | Naranja | j &lt;= valor | **Bucle interno** - condición de columnas |
| 🟧 **Rombo** | Naranja | i &lt;= Orden | **Condición 1:** ¿Estamos en la mitad superior? |
| 🟧 **Rombo** | Naranja | j &lt;= k | **Condición 2:** ¿Posición válida para imprimir? |
| 🟧 **Rombo** | Naranja | i &lt;= Orden | **Condición 3:** (rama alternativa) ¿Mitad superior? |
| 🟧 **Rombo** | Naranja | j &lt;= k | **Condición 4:** ¿Posición válida en rama alternativa? |
| 🟩 **Paralelogramo** | Verde | Salida "i" & valor-1+j & "j" & i-Orden+1 & "k" & k | **Salida A:** Coordenadas mitad superior |
| 🟩 **Paralelogramo** | Verde | Salida "i" & valor-1+j & "j" & Orden-i+1 & "k" & k | **Salida B:** Coordenadas mitad inferior |
| 🟩 **Paralelogramo** | Verde | Salida "i" & valor-j+1 & "j" & valor-1 & "k" & k | **Salida C:** Coordenadas izquierda mitad superior |
| 🟩 **Paralelogramo** | Verde | Salida "i" & valor-j+1 & "j" & valor & "k" & k | **Salida D:** Coordenadas izquierda mitad inferior |
| 🟨 **Rectángulo** | Amarillo | k = k+1 | Incrementar ciclo externo |
| 🟨 **Rectángulo** | Amarillo | j = j+1 | Incrementar columna |
| 🟨 **Rectángulo** | Amarillo | i = i+1 | Incrementar fila |
| 🔵 **Óvalo** | Morado | Fin | Fin del programa |
```

```c++

## Fórmulas Matemáticas Clave

| Fórmula | Propósito |
|---------|-----------|
| `n = Orden * 2 - 1` | Calcula dimensión total de la matriz (siempre impar) |
| `valor = n/2 - Abs(Int(n/2 + 0.5) - i) + 1` | Calcula ancho de la fila i (patrón diamante) |
| `i - Orden + 1` | Coordenada relativa en mitad superior |
| `Orden - i + 1` | Coordenada relativa en mitad inferior |
| `valor - 1 + j` | Posición horizontal derecha |
| `valor - j + 1` | Posición horizontal izquierda |

## Flujo del Algoritmo (Paso a Paso)

| Paso | Acción | Descripción |
|------|--------|-------------|
| 1 | Inicio | Comienza el programa |
| 2 | Declarar variables | i, j, k, Orden, valor, n |
| 3 | Pedir "Orden" | Solicitar tamaño de la matriz |
| 4 | Leer Orden | Input del usuario |
| 5 | Calcular n | `n = Orden * 2 - 1` (dimensión total) |
| 6 | Inicializar k = 1 | Contador de ciclos externos |
| 7 | Inicializar i = 1 | Contador de filas |
| 8 | ¿i <= n? | Verificar fin de filas |
| 9 | Calcular valor | Ancho de la fila actual |
| 10 | Inicializar j = 1 | Contador de columnas |
| 11 | Mostrar "Ciclo externo:" & k | Información del ciclo |
| 12 | ¿j <= valor? | Verificar fin de columnas |
| 13 | ¿i <= Orden? | ¿Mitad superior o inferior? |
| 14 | ¿j <= k? | ¿Derecha o izquierda? (mitad sup) |
| 15 | Mostrar coordenadas A o C | Según condición anterior |
| 16 | ¿j <= k? | ¿Derecha o izquierda? (mitad inf) |
| 17 | Mostrar coordenadas B o D | Según condición anterior |
| 18 | k = k + 1 | Incrementar ciclo externo |
| 19 | j = j + 1 | Siguiente columna |
| 20 | i = i + 1 | Siguiente fila |
| 21 | Fin | Programa termina |

## Ejemplo con Orden = 3 (n = 5)

| i (fila) | valor | Mitad | j recorrido | Coordenadas generadas |
|----------|-------|-------|-------------|----------------------|
| 1 | 1 | Superior | 1 | Centro superior |
| 2 | 2 | Superior | 1, 2 | Derecha e izquierda |
| 3 | 3 | Centro | 1, 2, 3 | Completa |
| 4 | 2 | Inferior | 1, 2 | Derecha e izquierda |
| 5 | 1 | Inferior | 1 | Centro inferior |
```
