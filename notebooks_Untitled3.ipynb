```c++
#include <iostream>
#include <sstream>
#include <string>
#include <cstdlib>
#include <cmath>

using namespace std;

// Headers
string toString (double);
int toInt (string);
double toDouble (string);

int main() {
    int i;

    i = 20;
    while (i <= 30) {
        cout << i << endl;
        i = i + 1;
    }
    return 0;
}

// The following implements type conversion functions.
string toString (double value) { //int also
    stringstream temp;
    temp << value;
    return temp.str();
}

int toInt (string text) {
    return atoi(text.c_str());
}

double toDouble (string text) {
    return atof(text.c_str());
}

```

```c++
1. Inicio (Principal)
   ↓
2. Declarar variable entera "i"
   ↓
3. Asignar i = 20
   ↓
4. ¿i <= 30? (Condición)
   ├── FALSO → Fin del programa
   ↓
   └── VERDADERO → Mostrar i (Salida i)
                     ↓
                     Incrementar i = i + 1
                     ↓
                     Volver al paso 4
```
