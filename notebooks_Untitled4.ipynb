```c++
1. Inicio (Principal)
   ↓
2. ¿true? (Condición)
   └── SIEMPRE VERDADERO → Mostrar "hola mundo"
                           ↓
                           Volver al paso 2 (bucle infinito)
   └── Falso → Nunca se ejecuta → Fin (inalcanzable)
```

```c++
#include <iostream>
#include <sstream>
#include <string>
#include <cstdlib>
#include <cmath>

using namespace std;

// Headers
string toString (double);
int toInt (string);
double toDouble (string);

int main() {
    while (true) {
        cout << "hola mundo" << endl;
    }
    return 0;
}

// The following implements type conversion functions.
string toString (double value) { //int also
    stringstream temp;
    temp << value;
    return temp.str();
}

int toInt (string text) {
    return atoi(text.c_str());
}

double toDouble (string text) {
    return atof(text.c_str());
}
```
